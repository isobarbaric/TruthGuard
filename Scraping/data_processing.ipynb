```python
science_url = 'https://mediabiasfactcheck.com/pro-science/'
conspiracy_url = 'https://mediabiasfactcheck.com/conspiracy/'
```

```python
from article_finder import ArticleFinder
article_compiler = ArticleFinder()
```

```python
import json
def write_to_storage(name: str, articles: list[str]):
    filename = "../Data/" + name + ".json" 
    with open(filename, 'w') as storage:
        storage.write(json.dumps(articles, indent = 4))
```

```python
science_articles = article_compiler.find_articles(science_url)
```

```python
write_to_storage('science', science_articles)
```

```python
conspiracy_articles = article_compiler.find_articles(conspiracy_url)
```

```python
write_to_storage('conspiracy', conspiracy_articles)
```

```python
min_len = min(len(science_articles), len(conspiracy_articles))
science_articles = science_articles[:min_len]
conspiracy_articles = conspiracy_articles[:min_len]
```

```python
import pandas as pd
import numpy as np
```

```python
df_science = pd.DataFrame.from_dict(science_articles)
df_conspiracy = pd.DataFrame.from_dict(conspiracy_articles)
```

```python
df_science['article_type'] = 1
df_conspiracy['article_type'] = 0
```

```python
# merge the two dataframes
df = pd.concat([df_science, df_conspiracy], axis=0, ignore_index=True)
```

```python
split_df = np.array_split(df, 30)
```

```python
from newspaper import Article
import time
```

```python
def parse_chunk(articles):
    start = time.time()
    extracted_text = []
    for article in articles:
        # print("parsing:", article[1])
        try:
            current_article = Article(article[1])
            current_article.download(), current_article.parse()
            extracted_text.append(current_article.text)
        except Exception:
            print("error occurred:", article[1])
    end = time.time()    
    print(end-start, "seconds elapsed")
    return extracted_text
```

```python
len(split_df)
```

```python
all_text = []
```

```python
# third_chunk = parse_chunk(split_df[2].values)
for chunk in split_df:
    parsed_current_chunk = parse_chunk(chunk.values)
    all_text.append(parsed_current_chunk)
```
